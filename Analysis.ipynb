## Dealing with categorical data

```python
data.dtypes
```

```python
data['diagnosis'] = data['diagnosis'].map({'B':0,'M':1})
```

## Making correlation and heatmap

```python
correlation = data.corr()['diagnosis'].sort_values(ascending=False)
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(correlation.to_frame(),annot=False,cmap='coolwarm',cbar=False,fmt='.2f')
plt.title('Correlation matrix')
plt.show()
```

```python
data.corrwith(data['diagnosis']).plot.bar(figsize=(20,10), title='correlated with diagnosis',rot=45, grid=True)
```

## Making new dataset with selected features

```python
selected_features = correlation[correlation >= 0.6].index.tolist()
selected_features.remove('diagnosis')
```

```python
new_data = data[selected_features + ['diagnosis']]
```
